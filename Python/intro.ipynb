```python
def calcMeans(dataframe):
    # Returns a NumPy ndarray instead of pandas Series
    means = dataframe.apply(lambda col: col.mean()).to_numpy()
    return means

```

```python
from sklearn import datasets
import pandas as pd
import numpy as np
iris_X, iris_y = datasets.load_iris(return_X_y=True, as_frame=True)
means = calcMeans(iris_X)
print("Calculated Means:\n", means)
print("\nValidation with describe():\n", iris_X.describe().loc['mean'])

```
